```python
import pandas as pd
df = pd.read_excel("/Users/varunpasad/Downloads/opening.xlsx")

df = df[df.SKU.notnull()]
df = df[df.Opening.notnull()]
df = df[-df.SKU.str.startswith("MRT_")]
df
```

```python
# odoo connn
import xmlrpclib
import datetime

DEFAULT_SERVER_DATE_FORMAT = "%Y-%m-%d"
DEFAULT_SERVER_TIME_FORMAT = "%H:%M:%S"
DEFAULT_SERVER_DATETIME_FORMAT = "%s %s" % (
    DEFAULT_SERVER_DATE_FORMAT,
    DEFAULT_SERVER_TIME_FORMAT)

username = 'varun@tendercuts.in' #the user
pwd = 'qwerty123'      #the password of the user
dbname = 'tcuts'    #the database

# Get the uid
sock_common = xmlrpclib.ServerProxy ('http://localhost:8071/xmlrpc/2/common')
uid = sock_common.login(dbname, username, pwd)
print (uid)
# replace localhost with the address of the server
sock = xmlrpclib.ServerProxy('http://localhost:8069/xmlrpc/2/object')
```

```python
# Santize sku names
products = sock.execute_kw(dbname, uid, pwd, 'product.product', 'search_read', [[]], {'fields': ["default_code"]})
for product in products:
    try:
        print(product)
        sock.execute_kw(dbname, uid, pwd, 'product.product', 'write',
                        [
                            product['id'],
                            {'default_code': product['default_code'].strip().replace(" ", "")}
                        ]
                       )
    except:
        print (product, "Failed")
```

```python
partner_id = sock.execute_kw(dbname, uid, pwd, 'res.partner', 'search', [
    [['name', '=', "Suguna"]]
])

# product.product_uom_kgm


order_lines = []
for index, row in df.iterrows():
    product_id = sock.execute_kw(dbname, uid, pwd, 'product.product', 'search_read', [
        [['default_code','=', "RM_" + row['SKU']]]
    ], {'fields': ['name', 'default_code'], 'limit': 1 })
    
    if (len(product_id) == 0):
        print (row['SKU'], "Failed")
        continue
        
    print (row['SKU'])
    product_id = product_id[0]
    
    order_lines.append((0, 0, {
        'name': product_id['name'],
        'product_id': product_id['id'],
        'price_unit': 100,
        'product_qty': row['Opening'],
        'date_planned': datetime.datetime.today().strftime(DEFAULT_SERVER_DATETIME_FORMAT),
        'product_uom': 3
        
    }))

po_vals = {
    'partner_id': partner_id[0],
    'order_line': order_lines
}


keck = sock.execute_kw(dbname, uid, pwd, "purchase.order", 'create', [po_vals])

```
