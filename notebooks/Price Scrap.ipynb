```python
# F2H
import requests

product_url = "https://www.freshtohome.com/xmlconnect/product/getAllProducts"
res = requests.get(product_url)
master_data = res.json()

price_url = "https://www.freshtohome.com/xmlconnect/product/getAvailableProducts"
res = requests.get(price_url)
price_master = res.json()
price_master
```

```python
import pandas as pd
data = [(product['name'], product['entity_id']) for product in master_data['products']]
master_df = pd.DataFrame(data, columns=['name', 'entity_id'])

data = [(product['final_price'], product['entity_id']) for product in price_master['products']]
price_df = pd.DataFrame(data, columns=['price', 'entity_id'])

# master_df.merge(price_df, on="entity_id").to_csv('/Users/varunpasad/Desktop/fresh.csv', encoding="utf-8")
price_df
```

```python
from bs4 import BeautifulSoup
# pip install "requests[security]"
import requests

session = requests.session()

# extract token
data = session.get("https://m.licious.in/chicken")
soup = BeautifulSoup(data.content)
csrf_token = soup.find('meta',{"name": "_token"})['content']

licious_master = []
for cat_id in [1, 2, 3]:
    data = session.post(
        "https://m.licious.in/checkout_oos/get-products",
        data={"cat_id": cat_id},
        headers={
            "X-CSRF-TOKEN": csrf_token,
            "X-Requested-With":"XMLHttpRequest",
            "Referer": "https://m.licious.in/chicken",
            "Origin":"https://m.licious.in",
            'User-Agent': 'Mozilla/5.0'
        })
    for product in data.json()['data']:
        licious_master.append((
            product['product_master']['product_id'],
            product['product_master']['pr_name'],
            product['product_master']['gross'],
            product['product_pricing']['base_price']
        ))
```

```python
pd.DataFrame(licious_master, columns=['id', 'name', 'gross', 'price']).to_csv("/Users/varunpasad/Desktop/licous.csv")
```

```python
import requests
bb_database = "https://www.bigbasket.com/mapi/v3.4.0/product/list/?page={}&slug=fresho-meat&sorted_on=&store_slug=&tab_type=%5B%22all%22%5D&type=pbpc"
session = requests.Session()
# acquire cookies
data = session.get("https://www.bigbasket.com/pb/fresho-meat/?nc=fapd")

i = 1
page = session.get(bb_database.format(i))
info = page.json()['response']['tab_info'][0]['product_info']
tot_pages = info['tot_pages']
master = []
while i <= int(tot_pages):
    print ("Running Page {}".format(i))
    if i != 1:
        page = session.get(bb_database.format(i))
        info = page.json()['response']['tab_info'][0]['product_info']

    for product in info['products']:
        master.append((
            product['sku'],
            product['p_desc'],
            product['sp'],
            product['w']
        ))
    
    # next page
    i += 1
```

```python
import pandas as pd
pd.DataFrame(master, columns=['id', 'name', 'price', 'weight']).to_csv("/Users/varunpasad/Desktop/bb.csv")
```
