```python
source_file = "/Users/varunpasad/projects/odoo/tcuts/data/hr.employee.csv"
images_dir = "/Users/varunpasad/projects/odoo/tcuts/data/emp_images/"
image_file = "/Users/varunpasad/projects/odoo/tcuts/data/hr.employee.image.csv"
```

```python
# load the sheet
import pandas as pd
import numpy as np
import os
import glob
import base64

        
class EmpImageGen:

    def __init__(self, df):
        self.df = df
    
    
    
    def build_row(self, row):
        """Generate the row data
        such as salelable, pruchasable etc
        """
        image_file = os.path.join(images_dir, "{}.JPG".format(row.identification_id))
        if os.path.exists(image_file):
            print(image_file)
            with open(image_file, "rb") as fh:
                row['image'] = base64.b64encode(fh.read())
            

        return row

    
    def to_csv(self, path):
        self.df = self.df.apply(self.build_row, axis=1)
        self.df.reset_index().to_csv(
            image_file,
            index=False,
            columns=["id", "image"]
            )
df = pd.read_csv(source_file)[['id', 'identification_id']]
EmpImageGen(df).to_csv(image_file)
```
