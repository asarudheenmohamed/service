```python
from bs4 import BeautifulSoup
import re
import random
import dateutil
import calendar

month = 2
year = 2018

with open("/Users/varunpasad/Desktop/act_template/vasant.html") as fd:
    soup = BeautifulSoup(fd)

#### Invoice number
element = soup.find_all(text=re.compile(r'Invoice No. :'))
# generate new one
bill_no = random.randint(00000001, 9999999)
new_no = re.sub(
    r"(.*)TN171(.*)",
    r"\1TN171%s" % bill_no,
    unicode(element[0])
)
element[0].replace_with(new_no)

#### Invoice date
element = soup.find_all(text=re.compile(r'Invoice Date :'))
inv_date = dateutil.parser.parse("01/{}/{}".format(month, year), dayfirst=True)
element[0].replace_with("Invoice Date :" + format(inv_date, "%d/%m/%y"))

### invoice period
element = soup.find_all(text=re.compile(r'Invoice Period :'))
prev_month = month - 1
from_day, to_day = calendar.monthrange(year, prev_month)
from_date = dateutil.parser.parse("01/{}/{}".format(prev_month, year), dayfirst=True)
from_date = format(from_date, "%d/%m/%y")
print (from_date)

to_date = dateutil.parser.parse("{}/{}/{}".format(to_day, prev_month, year))
to_date = format(to_date, "%d/%m/%y")
element[0].replace_with("Invoice Period : {} - {}".format(from_date, to_date))

### due data
element = soup.find_all(text=re.compile(r'Due Date :'))
due_date = dateutil.parser.parse("14/{}/{}".format(month, year), dayfirst=True)
element[0].replace_with("Due Date : " + format(due_date, "%d/%m/%y"))

### txn data
element = soup.find_all(text=re.compile(r'01-JAN-18'))
due_date = dateutil.parser.parse("01/{}/{}".format(month, year), dayfirst=True)
element[0].replace_with(format(due_date, "%d-%b-%y"))

### period
element = soup.find_all(text=re.compile(r'01/12/2018-31/12/2018'))
prev_month = month - 1
from_day, to_day = calendar.monthrange(year, prev_month)
from_date = dateutil.parser.parse("01/{}/{}".format(prev_month, year), dayfirst=True)
from_date = format(from_date, "%d/%m/%y")

to_date = dateutil.parser.parse("{}/{}/{}".format(to_day, prev_month, year), dayfirst=True)
to_date = format(to_date, "%d/%m/%y")
element[0].replace_with("{} - {}".format(from_date, to_date))

## txn number
element = soup.find_all(text=re.compile(r'10322259683'))
bill_no = random.randint(00000001, 99999999)
element[0].replace_with("103"+ str(bill_no))

### quantity
element = soup.find_all(text=re.compile(r'31 Days'))
from_day, to_day = calendar.monthrange(year, prev_month)
element[0].replace_with("{} Days".format(to_day))


with open("/Users/varunpasad/Desktop/act_template/output1.html", "w") as file:
    file.write(str(soup))
```
